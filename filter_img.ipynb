```python
from facenet_pytorch import MTCNN
from PIL import Image
import os
```

```python
# Thư mục chứa ảnh đã sinh
input_dir = "/content/drive/MyDrive/Work_space/project_facial_expression/stylegan2_fer2013/generate/generated_images"
output_dir = "/content/drive/MyDrive/Work_space/project_facial_expression/stylegan2_fer2013/generate/filtered_images"
os.makedirs(output_dir, exist_ok=True)

detector = MTCNN(keep_all=False, device='cuda')

for fname in os.listdir(input_dir):
    path = os.path.join(input_dir, fname)
    try:
        img = Image.open(path).convert('RGB')
        boxes, _ = detector.detect(img)
        if boxes is not None and len(boxes) > 0:
            img.save(os.path.join(output_dir, fname))
    except Exception as e:
        print(f"Skip {fname}: {e}")
```
